Проводим обучение нейронной сети для распознавания свободного и занятого места на парковке

```
from imageai.Detection.Custom import DetectionModelTrainer

trainer = DetectionModelTrainer()
trainer.setModelTypeAsYOLOv3()
trainer.setDataDirectory(data_directory="/content/drive/MyDrive/lenina")
trainer.setTrainConfig(object_names_array=["empty_spot", "occupied_spot"], batch_size=4, num_experiments=100, train_from_pretrained_model="/content/drive/MyDrive/headsets/pretrained-yolov3.h5")
trainer.trainModel()
```

Находим наилучшую модель по методу **mAP**

```
from imageai.Detection.Custom import DetectionModelTrainer

trainer = DetectionModelTrainer()
trainer.setModelTypeAsYOLOv3()
trainer.setDataDirectory(data_directory="/content/drive/MyDrive/lenina")
metrics = trainer.evaluateModel(model_path="/content/drive/MyDrive/lenina/models", json_path="/content/drive/MyDrive/lenina/json/detection_config.json", iou_threshold=0.5, object_threshold=0.3, nms_threshold=0.5)
print(metrics)
```

Проводим тестовый запуск на картинке с одной парковкой

```
from imageai.Detection.Custom import CustomObjectDetection
from IPython.display import Image

detector = CustomObjectDetection()
detector.setModelTypeAsYOLOv3()
detector.setModelPath("/content/drive/MyDrive/lenina/models/detection_model-ex-009--loss-0026.142.h5")
detector.setJsonPath("/content/drive/MyDrive/lenina/json/detection_config.json")
detector.loadModel()

detections = detector.detectObjectsFromImage(input_image="/content/drive/MyDrive/lenina/test.jpg", 
                                             output_image_path="/content/drive/MyDrive/lenina/output_test.jpg", 
                                             minimum_percentage_probability=43,
                                             nms_treshold=0.4)

for detection in detections:
    print(detection["name"], " : ", detection["percentage_probability"], " : ", detection["box_points"])

Image('/content/drive/MyDrive/lenina/output_test.jpg')
```

Проводим испытание на отрезке видео, где машина заезжает на парковку

```
from imageai.Detection.Custom import CustomVideoObjectDetection
import os

execution_path = os.getcwd()

video_detector = CustomVideoObjectDetection()
video_detector.setModelTypeAsYOLOv3()
video_detector.setModelPath("/content/drive/MyDrive/lenina/models/detection_model-ex-009--loss-0026.142.h5")
video_detector.setJsonPath("/content/drive/MyDrive/lenina/json/detection_config.json")
video_detector.loadModel()


def forFull(output_arrays, count_arrays, average_output_count):
    print("Array for the outputs of each frame ", output_arrays)
    print("Array for output count for unique objects in each frame : ", count_arrays)
    print("Output average count for unique objects in the entire video: ", average_output_count)
    print("------------END OF THE VIDEO --------------")

video_detector.detectObjectsFromVideo(input_file_path="/content/drive/MyDrive/lenina/lenina_car_out.mp4",
                                        output_file_path=os.path.join(execution_path, "/content/drive/MyDrive/lenina/lenina_car_out_detect"),
                                        frame_detection_interval=1,
                                        frames_per_second=25,
                                        minimum_percentage_probability=50,
                                        log_progress=True,
                                        video_complete_function=forFull)
```
